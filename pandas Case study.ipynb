```python
#1. which team played the most fifa world cup matches
# we concatenate (combine) the home and away teams and do a value counts
```

```python
pd.concat([df['home__team'],df['away__team']]).value_counts()
```

```python
# highest scoring tournaments
# group by year
#aggregate (sum of total goals)
#sort descending
#filter top 5
yearly_goals = df.groupby('year')['total_goals'].sum().sort_values(ascending=False)
yearly_goals
```

```python
top5_years = yearly_goals[:5]
top5_years
```

```python
top5_years.index
```

```python
## which countries hosted the world cup in these years
#filter for the top scoring years
#look for the host team is True
#drop duplicates
```

```python
condition1 = df['year'].isin(top5_years.index)
condition2 = df['host__team'] == True
df[condition1 & condition2][['year', 'home__team','host__team' ]].drop_duplicates()
```

```python
#3 How many goals has each team scored in the fifa world cup? which team has scored 
# approach is to get  all the gosls scored by each country as home team;
# get all goals scored by each country as away team;
#concatenate (combine) the two arrays from step 1 and 2
# then get the total goals for each country
 
```

```python
# total goals as home and away
df.groupby('home__team') ['hg'].sum().reset_index()
home_goals_df = df.groupby('home__team') ['hg'].sum().reset_index()
aways_goals_df = df.groupby('away__team') ['ag'].sum().reset_index()
```

```python
aways_goals_df
```

```python
home_goals_df
```

```python
#combine the two arrays
total_goals_df = pd.concat([home_goals_df.rename(columns={'home__team':'team','hg':'goals'}),
           aways_goals_df.rename(columns={'away__team': 'team','ag':'goals'})],ignore_index=True)
```

```python
total_goals_df.groupby('team')['goals'].sum().sort_values(ascending=False)
```
